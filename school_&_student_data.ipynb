```python
import pandas as pd


df2019 = pd.read_csv("Resources/Cohort_2019.csv")
df2020 = pd.read_csv("Resources/Cohort_2020.csv")
df2021 = pd.read_csv("Resources/Cohort_2021.csv")
df2022 = pd.read_csv("Resources/Cohort_2022.csv")
df2023 = pd.read_csv("Resources/Cohort_2023.csv")
```

```python
df_combined = pd.concat([df2019, df2020, df2021, df2022, df2023], ignore_index=True)
df_combined
```

```python
df_combined = df_combined.set_index("Cohort Year")
df_combined.tail()
```

```python
df_combined = df_combined.drop(columns=['Graduation Rate Type'])
df_combined

```

```python
df_combined.columns
```

```python
# Extract unique values from a specific column
df_combined['Subgroup'].unique

```

```python
# subgroups_df= df_combined[df_combined["Subgroup"]== "Male","Female", "Hispanic or Latino", "Homeless", "All"]

subgroups_df = df_combined[df_combined['Subgroup'].isin(["Male", "Female", "Hispanic or Latino", "Homeless", "All"])]

subgroups_df
```

```python
subgroups_df['Subgroup'] = subgroups_df['Subgroup'].replace('Homeless', 'Unhoused')
```

```python
subgroups_df.to_csv('merged_cohorts_df.csv', index=True)  # 'index=False' if we decided to exclude it from the CSV
```
